¿Cuál es el registro de personas extraviadas o desaparecidas en México durante el periodo comprendido de 1968 – 2018?

```python
import pandas as pd
import numpy as np
```

```python
df= pd.read_csv('https://raw.githubusercontent.com/MinervaNunez/Proyecto-de-Python-BEDU/Rama-Jesus/RNPEDFC.csv', sep=',', encoding='ISO-8859-1')
df
```

```python

fin=len(df.index)
fin
```

```python
lst = list(np.arange(1,fin+1))
#lst
len(lst)
```

```python
df=df.reindex(lst, method='ffill')
df
```

```python
column_name_mapping = {
    'Fecha en que se le vio por ultima vez': 'fecha_ultvz',
    'Hora en que se le vio por ultima vez': 'hora_ultvz',
    'Pais en que se le vio por ultima vez': 'pais_ultvz',
    'Entidad en que se le vio por ultima vez': 'entidad_ultvz',
    'Municipio en que se le vio por ultima vez': 'municipio_ultvz',
    'Localidad en que se le vio por ultima vez': 'localidad_ultvz',
    'Nacionalidad': 'nacionalidad',
    'Estatura': 'estatura',
    'Complexion': 'complexion',
    'Sexo': 'sexo',
    'Edad': 'edad',
    'Descripcion de senas particulares': 'senas_particulares',
    'Etnia': 'etnia',
    'Discapacidad': 'discapacidad',
    'Dependencia que envio la informacion': 'dependecia_origen'
    
}
```

```python
df_2 = df.rename(columns=column_name_mapping)

df_2
```

```python
df_2.dtypes
```

```python
print(f'{("Columna"):42} =NO ESPECIFICADO= \n')
for col in df_2.columns:
    print(f'{(col):42} | {(df_2[col] == "NO ESPECIFICADO").sum()}')
```

```python
# Quitamos los no especificados del campo fecha, colocando en su lugar otra cadena =0000-00-00
df_2.loc[df_2['fecha_ultvz'] == 'NO ESPECIFICADO', 'fecha_ultvz'] = '0000-00-00'
#Revisamos de nuevo 
print(f'{("Columna"):42} =NO ESPECIFICADO= \n')
for col in df_2.columns:
    print(f'{(col):42} | {(df_2[col] == "NO ESPECIFICADO").sum()}')
```

```python
#Convertimos el tipo de dato de fecha_ultvz especificando que si hay un error que continue
df_2['fecha_ultvz']= pd.to_datetime(df_2['fecha_ultvz'],errors='coerce')

#Convertimos la serie de edad a numerico con pd.to_numeric, inidicando que si encuentra error que continue, pero los pone como NAN
df_2['edad']= pd.to_numeric(df_2['edad'],errors='coerce')

#rellenamos los NAN de Edad con ceros
df_2["edad"]=df_2["edad"].fillna(0)
#Ahora si convertimos la columna edad a int
df_2["edad"]= df_2["edad"].astype(int)
df_2.dtypes
```
